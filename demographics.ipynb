```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
import pingouin as pg
import seaborn as sns
```

```python
df = pd.read_csv('vhv_11_15.csv')
```

```python
df['age'].describe()
```

```python
def assess_var(var):
    skew = var.skew()
    kurt = var.kurtosis()
    return f'The skewness is {skew} and the kurtosis is {kurt}.'
assess_var(df['age'])
```

```python
df['age'].plot(kind='hist', bins = 65, figsize=(8, 5))

# add a title to the histogram
pyplot.title('Age frequencies')
# add y-label
pyplot.ylabel('Frequencies')
# add x-label
pyplot.xlabel('Age')

pyplot.show()
```

```python
df['age'].plot(kind='box', figsize=(8, 5))

# add a title to the histogram
pyplot.title('Age frequencies')
# add y-label
pyplot.ylabel('Frequencies')
# add x-label
pyplot.xlabel('Age')

pyplot.show()
```
